```python
#open csv file siero_v2.csv in a pandas dataframe
import pandas as pd
import numpy as np

dfsiero = pd.read_csv('siero_v2.csv', sep=',', header=0, index_col=0)

dfsiero
```

```python
#translate row in prolog facts for each attribute for example for the first row newpatient is patient(it_0154_004787).
#print the facts in a separate prolog file

with open('facts2.pl', 'w') as f:
    #for each row of the dfsiero dataframe (each patient)
    for index, row in dfsiero.iterrows():
        #for each column of the row (each attribute)
        for col in dfsiero.columns:
            #if the value of the attribute is not NaN
            if not pd.isnull(row[col]):
                #print the fact in the file
                f.write(col + '(' + index + ',' + str(row[col]) + ').\n')




```

```python
#take the file facts2.pl and make it all lowercase
import re
with open('facts2.pl', 'r') as f:
    with open('facts2lower.pl', 'w') as g:
        for line in f:
            g.write(line.lower())

#take the file facts2lower.pl and remove the spaces between the words and instead put an underscore
with open('facts2lower.pl', 'r') as f:
    with open('facts2lowerunderscore.pl', 'w') as g:
        for line in f:
            g.write(re.sub(r' ', '_', line))

#if there are '-' in the file facts2lowerunderscore.pl remove them and put an underscore instead
with open('facts2lowerunderscore.pl', 'r') as f:
    with open('facts2lowerunderscore2.pl', 'w') as g:
        for line in f:
            g.write(re.sub(r'-', '_', line))
        
```

```python
#do the same translation from the dfsiero dataframe in prolog facts but translate only the attributes that are in the list
#of the attributes to translate
listofacto = ['Gender', 'Race', 'numberdoses', 'mobcat', 'PA3mcat', 'totndrugs', 'frailty01', 'CKD01', 'poornut01', 'cancer01', 'immunedis01', 'IBD01', 'liverdis01', 
              'DM01_tot', 'OAD01_tot', 'HTA01_tot', 'CVD01_tot', 'AF01_tot', 'PAD01_tot', 'chrrespdis01_tot', 'obesity01_tot', 'depress01_tot',
              'cogdis01_tot', 'park01_tot', 'thyrdis01_tot', 'epilepsia01_tot', 'anxiety01_tot', 'HD01_tot', 'urol01_tot', 'Mmsumrev', 'excovidtot', 'antiS_0', 'antiS_2', 
              'antiS_6', 'antiS_12', 'covid6m', 'covid12m', 'WHO_f1', 'WHO_f2', 'WHO_f3', 'WHO_f4', 'TypeofOutcome_updated']

with open('factsto.pl', 'w') as f:
 
    #for each row of the dfsiero dataframe (each patient)
    for index,row in dfsiero.iterrows():
        #for each column of the row (each attribute)
        for col in dfsiero.columns:
            #if the value of the attribute is not NaN and the attribute is in the list of the attributes to translate
            if not pd.isnull(row[col]) and col in listofacto:
                #print the fact in the file
                f.write(col + '(' + index + ',' + str(row[col]) + ').\n')
    #print a space after very patient
        f.write('\n')

with open('indfacts.pl', 'w') as f:
    i=0
    #for each row of the dfsiero dataframe (each patient)
    for index,row in dfsiero.iterrows():
        #for each column of the row (each attribute)
        for col in dfsiero.columns:
            #if the value of the attribute is not NaN and the attribute is in the list of the attributes to translate
            if not pd.isnull(row[col]) and col in listofacto:
                #print the fact in the file
                f.write(col + '(' + str(i) + ',' + str(row[col]) + ').\n')

            if col == 'TypeofOutcome_updated':
                if row[col] == 1:
                    f.write('patient(' + str(i) + ','+ 'pos'+').\n')
                elif row[col] == None:
                    f.write('patient(' + str(i) + ','+ 'unknown'+').\n')
                else:
                    f.write('patient(' + str(i) + ','+ 'neg'+').\n')
                    
    #print a space after very patient
        i += 1
        f.write('\n')

#take the file factsto.pl and make it all lowercase
import re
with open('indfacts.pl', 'r') as f:
    with open('factstolower.pl', 'w') as g:
        for line in f:
            g.write(line.lower())


#take the file factstolower.pl and remove the spaces between the words and instead put an underscore
with open('factstolower.pl', 'r') as f:
    with open('factstolowerunderscore.pl', 'w') as g:
        for line in f:
            g.write(re.sub(r' ', '_', line))

#if there are '-' in the file factstolowerunderscore.pl remove them and put an underscore instead
with open('factstolowerunderscore.pl', 'r') as f:
    with open('factstolowerunderscore2.pl', 'w') as g:
        for line in f:
            g.write(re.sub(r'-', '_', line))

#remove '01_tot' from the attributes in the file factstolowerunderscore2.pl
with open('factstolowerunderscore2.pl', 'r') as f:
    with open('factstoFINAL.pl', 'w') as g:
        for line in f:
            g.write(re.sub(r'01_tot', '', line))

```
