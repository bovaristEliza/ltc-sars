```python
#open csv file siero_v2.csv in a pandas dataframe
import pandas as pd
import numpy as np

#dfsiero = pd.read_csv('siero_v2.csv', sep=',', header=0, index_col=0)
dfsiero = pd.read_excel('/home/apsev/ltc-sars/Vax_v2.xlsx', sheet_name='Rawdata')
dfsiero
```

```python
#make missing val typevax and numberdoses as 9
dfsiero['typevax'] = dfsiero['typevax'].fillna(9)
dfsiero['numberdoses'] = dfsiero['numberdoses'].fillna(9)
dfsiero['TypeofOutcome_updated'] = dfsiero['TypeofOutcome_updated'].fillna(9)
```

```python
#remove the rows from dfsiero where all WHO_f1, WHO_f2, WHO_f3, WHO_f4 are equal to 9
dfsiero = dfsiero[(dfsiero.WHO_f1 != 9) | (dfsiero.WHO_f2 != 9) | (dfsiero.WHO_f3 != 9) | (dfsiero.WHO_f4 != 9)]
dfsiero
```

```python
import numpy as np

# Initialize an empty list to store the labels
labels = []
patient = []

# Iterate through the DataFrame using .iterrows() for row-wise operation
for index, row in dfsiero.iterrows():
    if row['WHO_f1'] == 0 and row['WHO_f2'] == 0 and row['WHO_f3'] == 0 and row['WHO_f4'] == 0:
        labels.append('N')
        patient.append(row['Newpatient'])
    elif ((row['WHO_f1'] == 0 or row['WHO_f2'] == 0 or row['WHO_f3'] == 0 or row['WHO_f4'] == 0) and (row['WHO_f1'] in [0, 9] and row['WHO_f2'] in [0, 9] and row['WHO_f3'] in [0, 9] and row['WHO_f4'] in [0, 9])):
        labels.append('N')
        patient.append(row['Newpatient'])
    elif row['WHO_f1'] in [1, 2, 3, 4, 5] or row['WHO_f2'] in [1, 2, 3, 4, 5] or row['WHO_f3'] in [1, 2, 3, 4, 5] or row['WHO_f4'] in [1, 2, 3, 4, 5]:
        labels.append('P')
        patient.append(row['Newpatient'])

print(labels)
print(patient)
```

```python
#for every element of list patient, make it lowercase and replace '-' with '_'
list2 = [x.lower().replace('-', '_') for x in patient]
print(list2)
```
