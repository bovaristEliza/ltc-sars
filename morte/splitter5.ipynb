```python
#open csv file siero_v2.csv in a pandas dataframe
import pandas as pd
import numpy as np

#dfsiero = pd.read_csv('siero_v2.csv', sep=',', header=0, index_col=0)
dfsiero = pd.read_excel('/home/apsev/ltc-sars/Vax_v2.xlsx', sheet_name='Rawdata')
dfsiero
```

```python
dfsiero['TypeofOutcome_updated'] = dfsiero['TypeofOutcome_updated'].fillna(9)
```

```python
#create a list where each element is either P (positive) or N (negative) depending on the value of the column 'TypeofOutcome_updated', if it's 1 then its a P, else a 0
list = []
for i in dfsiero['TypeofOutcome_updated']:
    if i == 1:
        list.append('P')
    else:
        list.append('N')

print(list)

#rename the list 'labels'
labels = list

```

```python
#create a list where each element is the index value but lowercase and with _ instead of -
list2 = []
for i in dfsiero['Newpatient']:
    list2.append(i.lower().replace('-', '_'))

print(list2)
```

```python
#now given the two lists list2 of the dataset and labels, stratify the dataset based on the labels
from sklearn.model_selection import train_test_split

list2_train, list2_test, labels_train, labels_test = train_test_split(list2, labels, test_size=0.3, stratify=labels)
```

```python
#count how many elements are there in list2_test and list2_train
print(len(list2_test))
print(len(list2_train))
```

```python
#count how many labels are positive and how many negative in labels_train and labels_test
print('TRAIN')
print(labels_train.count('P'))
print(labels_train.count('N'))
print('TEST')
print(labels_test.count('P'))
print(labels_test.count('N'))

#count the proportion of positive and negative labels in labels_train and labels_test
print('TRAIN')
print(labels_train.count('P')/len(labels_train))
print(labels_train.count('N')/len(labels_train))
print('TEST')
print(labels_test.count('P')/len(labels_test))
print(labels_test.count('N')/len(labels_test))

#count the proprotion of P and N in the original dataset
print('ORIGINAL')
print(labels.count('P')/len(labels))
print(labels.count('N')/len(labels))
```
